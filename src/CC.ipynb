```python
import dynfc as dyn
import numpy as np
from numpy.random import seed, rand
import scipy as sc
from scipy import io
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ts = sc.io.loadmat('data/ts.mat')['ts']
ts = ts.transpose()
```

```python
size = 24
corr_mats = dyn.corr_slide(ts, size)
corrcorr = dyn.cc(ts,24)
```

```python
mat1 = corr_mats[:,:,0]
iu1 = np.tril_indices(mat1.shape[0], k = 1)
mat1[iu1] = np.nan

aa = plt.figure(figsize = [4,4])
ax = sns.heatmap(mat1, 
                 cmap = "RdBu_r", 
                 vmin = -1, 
                 vmax = 1, 
                 #square = True, 
                 cbar = False)

ax.axis('off')
ax.tick_params(left=False, bottom=False)

aa.savefig('corr1_upper.pdf')
```
