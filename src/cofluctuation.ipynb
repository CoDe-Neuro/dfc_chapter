```python
import numpy as np
import scipy as sc
from scipy import io
import dynfc as dyn
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ts = sc.io.loadmat('data/ts.mat')['ts']
ts = ts.transpose()
ts.shape
```

```python
mat1 = dyn.corr_slide(ts,24)[:,:,0]

aa = plt.figure(figsize = [6,6])
ax = sns.heatmap(mat1, 
                 cmap = "RdBu_r", 
                 vmin = -1, 
                 vmax = 1, 
                 square = True, 
                 cbar = False)
ax.axis('off')
ax.tick_params(left=False, bottom=False)

aa.savefig('mat.pdf')
```

```python
mat1 = dyn.corr_slide(ts,24)[:,:,0]

aa = plt.figure(figsize = [6,6])
ax = sns.heatmap(mat1, 
                 cmap = "RdBu_r", 
                 vmin = -1, 
                 vmax = 1, 
                 square = False, 
                 cbar = True)
ax.axis('off')
ax.tick_params(left=False, bottom=False)

aa.savefig('mat_bar.pdf')
```

```python
mat1 = dyn.corr_slide(ts,24)[:,:,0]
iu1 = np.tril_indices(mat1.shape[0], k = -1)
mat1[iu1] = np.nan

aa = plt.figure(figsize = [6,6])
ax = sns.heatmap(mat1, 
                 cmap = "RdBu_r", 
                 vmin = -1, 
                 vmax = 1, 
                 #square = True, 
                 cbar = False)

ax.axis('off')
ax.tick_params(left=False, bottom=False)

aa.savefig('mat_upper.pdf')
```

```python
edges_series, rss = dyn.get_edgests(ts, 24)
```

```python
aa = plt.figure(figsize = (12/50,4))
ax = sns.heatmap(edges_series[:,0:1], 
                 cmap = "RdBu_r", 
                 vmin = -1, 
                 vmax = 1, 
                 square = False, 
                 cbar = False)
ax.axis('off')

aa.savefig('vec.pdf')
```
